```python
# Survival of Passengers based on gender and age

import altair as alt
alt.Chart(data).mark_bar().encode(
    alt.X('Sex'),
    alt.Y('count(PassengerId)', title='Number of Passengers'),
    color='Survived',
    column='Age'
)
```

```python
sns.countplot(x='Sex', hue='Survived', data=data)
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.show()
```
